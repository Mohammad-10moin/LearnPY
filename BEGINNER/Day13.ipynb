### Magic/Dunder method

```python
# __init__
# __len__
# __str__ goal: to be readable
# __repr__ goal: to be unambigious
# __call__
class A:
    def __init__(self,name):
        self.name=name
    def __len__(slef):
        return len(slef.name)
    def __str__(self):
        return f"it tells about the class "
    def __repr__(self):
        return f"it removes confusion"
    def __call__(self):
        return f"it calls the object"
a=A("mon")
print(a.name)
print(len(a))
print(len(a.name))
print(a)
print(str(a))
print(repr(a))
a()
```

### Method Overriding

```python
from math import pi
class Base:
    def meth(self):
        print("this is base method ")
class Area(Base):
    def __init__(self,x):
        self.radius=x
    def meth(self):
        print(f"The area of circle is {pi*self.radius*self.radius}")
c=Area(10)
c.meth()
```

### operator overloading

```python
class Vector:
    def __init__(self,x,y,z):
        self.i=x
        self.j=y
        self.k=z
    def __add__(self,x):
        return Vector(self.i+x.i,self.j+x.j,self.k+x.k)
    def __str__(self):
        return f" {self.i}i+{self.j}j+{self.k}k "
v1=Vector(1,2,3)
v2=Vector(3,2,1)
print(v1+v2)
```
